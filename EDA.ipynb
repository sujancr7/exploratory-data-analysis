Similar to the conditional expression, the isin() conditional function returns a True for each row the values are in
the provided list. To filter the rows based on such a function, use the conditional function inside the selection brackets
[]. In this case, the condition inside the selection brackets titanic["Pclass"].isin([2, 3]) checks for
which rows the Pclass column is either 2 or 3.

```python
#Simple way with OR operator
class_2_3 = titanic[(titanic["Pclass"] == 2) |
                    (titanic["Pclass"] == 3) ]
class_2_3.head()
```

# Handle time series data

```python
air_quality = air_quality.rename(columns={"date.utc":"datetime"})
air_quality.head()
air_quality["datetime"] = pd.to_datetime(air_quality["datetime"])
air_quality["datetime"]
```

```python
air_quality["datetime"].max(), air_quality["datetime"].min()
```

```python
#length of time series
air_quality["datetime"].max() - air_quality["datetime"].min()
```

```python
#add a new column to the DataFrame containing only the month of the measurement
air_quality.head()
```

```python
air_quality["month"]= air_quality["datetime"].dt.month
air_quality.head()
```

```python
air_quality["year"]= air_quality["datetime"].dt.year
air_quality.head()
```

```python
air_quality["weekofyear"]= air_quality["datetime"].dt.weekofyear
air_quality.head()
```

```python
air_quality["quarter"]= air_quality["datetime"].dt.quarter
air_quality.head()
```

```python
air_quality["weekday"]= air_quality["datetime"].dt.weekday
air_quality.head()
```

```python
#What is the average 𝑁𝑂2 concentration for each day of the week for each of the measurement locations?

average_NO2_each_day = air_quality.groupby([air_quality["datetime"].dt.weekday,
                                         "location"])["value"].mean()
average_NO2_each_day
```

what is the average value for each hour of the day?

```python
import matplotlib.pyplot as plt
```

```python
fig, axs = plt.subplots(figsize=(12,4))
air_quality.groupby(air_quality["datetime"].dt.hour)["value"].mean().plot(
kind='bar', rot=0, ax=axs)
plt.xlabel("Hour of the day");
plt.ylabel("$NO2(µg/m³)$");
```
